```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# Distribution Plot

```python
sns.distplot(tips['tip'],kde=False,bins=40,)
```

```python
sns.jointplot(tips['tip'],tips['total_bill'],data=tips,space=0.5,kind='hex')
```

```python
sns.barplot(tips['tip'],tips['total_bill'],data=tips,hue='sex',)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.scatterplot(x='tip',y='total_bill',data=tips,marker='+',y_jitter=True)
```

# Categorical plot

```python
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,hue='smoker')
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='sex',y='tip',data=tips,hue='smoker')
```

```python
sns.boxenplot('smoker','total_bill',data=tips,width=0.8)
```

```python
sns.violinplot('day','total_bill',data=tips,hue='smoker')
```

```python
sns.stripplot('day','tip',data=tips,hue='smoker',split=True)
```

```python
sns.swarmplot('day','tip',data=tips,hue='smoker',split=True)
```
