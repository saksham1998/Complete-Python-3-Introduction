```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
#Functional

plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('My Title')
```

```python
#Making multiple plots on same set
plt.subplot(2,3,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(y,x)

plt.subplot(2,2,3)
plt.plot(y**2,x**3)
```

```python
#OOP Method

fig = plt.figure()
axes = fig.add_axes([0.1,0.2,0.9,0.9]) 
axes2 = fig.add_axes([0.2,0.7,0.5,0.2])
#axes.plot(x,y)

axes.plot(x,y)
axes.set_title('Title 1')

axes2.plot(y,x)
axes.set_title('Title 2')
```
