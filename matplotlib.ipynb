```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
#Functional

plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('My Title')
```

```python
#Making multiple plots on same set
plt.subplot(1,3,1)
plt.plot(x,y)

plt.subplot(1,3,2)
plt.plot(y,x)

plt.subplot(1,3,3)
plt.plot(y**2,x**3)
```

```python
#OOP Method

fig = plt.figure()
axes = fig.add_axes([0.1,0.2,0.9,0.9]) 
axes2 = fig.add_axes([0.2,0.7,0.5,0.2])
#axes.plot(x,y)

axes.plot(x,y)
axes.set_title('Title 1')

axes2.plot(y,x)
axes.set_title('Title 2')
```

```python
# How to create subplots in OOP method:

fig,axes = plt.subplots(nrows=1,ncols=3)
plt.plot()
plt.tight_layout()#for better layout

axes[0].plot(x,y)
axes[0].set_title('Title 1')


axes[1].plot(y,x)
axes[1].set_title('Title 2')

axes[2].plot(y**2,x**3)
axes[2].set_title('Title 3')
```

# Figures and DPI

```python
fig = plt.figure(figsize=(8,2),dpi=100)#figsize is size of the figure in inches .DPI is Dots per Inch
ax = fig.add_axes([.1,.1,.9,.9])
ax.plot(x,y,'r')
```

```python
#or by subplot example

fig,axes = plt.subplots(figsize=(10,3),dpi=100)
axes.plot(y,x,'b')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(8,3))
a=x
b=y
for each in axes:
    each.plot(a,b)
    a,b = a**3,b**3   
plt.tight_layout()    
```
