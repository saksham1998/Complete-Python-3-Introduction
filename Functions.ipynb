```python
#Functions practise........................

#1. Write a function that returns the lesser of two given numbers if both numbers are even, but returns the greater if one or both numbers are odd

def my_func(a,b):
    if a%2==0 and b%2==0:
        return min(a,b)
    else:
        return max(a,b)
value = my_func(10,20)
print(value)
value = my_func(9,20)
print(value)

#2. Write a function takes a two-word string and returns True if both words begin with same letter

def my_func(a):
    arr = a.split()
    return (arr[0])[:1]==(arr[1])[:1]

value = my_func('sword sight')
print(value)
value = my_func('sword fight')
print(value)

#3. Given two integers, return True if the sum of the integers is 20 or if one of the integers is 20. If not, return False

def my_func(a,b):
    if (a+b==20) or (a==20) or (b==20):
        return True
    else:
        return False

value = my_func(11,9)
print(value)
value = my_func(20,9)
print(value)
value = my_func(11,11)
print(value)

#4. Write a function that capitalizes the first and fourth letters of a name

def my_func(a):
    return a[0].upper()+a[1:3]+a[3].upper()+a[4:]
value = my_func('saksham')
print(value)
value = my_func('helloworld')
print(value)

#5. Given a sentence, return a sentence with the words reversed

def my_func(sentence):
    arr = sentence.split()
    arr.reverse()
    return ' '.join(arr)
    
value = my_func('hello world')
print(value)
value = my_func('hello world how you doin')
print(value)

#6. Given an integer n, return True if n is within 10 of either 100 or 200

def my_func(num):
    if abs(100-num)<=10 or abs(200-num)<=10:
        return True
    else:
        return False

value = my_func(109)
print(value)
value = my_func(159)
print(value)
value = my_func(199)
print(value)

#7. Given a list of ints, return True if the array contains a 3 next to a 3 somewhere.

def my_func(arr):
    str1 = ''.join(str(e) for e in arr)
    if '33' in str1:
        print(True)

my_func([1,3,3])    
my_func([1,3,1,3])    

#8.Given a string, return a string where for every character in the original there are three characters

def my_func(string):
    mystring = ''
    for every in string:
        every = every*3
        mystring+=every
    print(mystring)    
my_func('hello')    
my_func('Mississippi')

#9. Given three integers between 1 and 11, if their sum is less than or equal to 21, return their sum. If their sum exceeds 21 and there's an eleven, reduce the total sum by 10. Finally, if the sum (even after adjustment) exceeds 21, return 'BUST'

def my_func(*args):
    
    if sum(args)<=21:
        print(sum(args))
    elif sum(args)>21 and 11 in args:
        x=sum(args)
        x-=10
        if x<21:
            print(x)
    else:
        print('BUST')
  
my_func(1,10,4)
my_func(11,2,9)
my_func(9,9,9)

#10. Return the sum of the numbers in the array, except ignore sections of numbers starting with a 6 and extending to the next 9 (every 6 will be followed by at least one 9). Return 0 for no numbers.

def my_func(arr):
    if 6 in arr and 9 in arr:
        first = arr.index(6)
        second = arr.index(9)
        if second>first:
            arr = arr[0:first]+arr[second+1:]
            print(sum(arr))
    else:
        print(sum(arr))
        
my_func([1, 3, 5])
my_func([2, 1, 6, 9, 11])
my_func([4, 5, 6, 7, 8, 9])

#11.  Write a function that takes in a list of integers and returns True if it contains 007 in order

def my_func(arr):
    #str1=''
    if 0 in arr and arr.count(0)>=2 and 7 in arr:
        str1 = ''.join(str(e) for e in arr)
        if '007' in str1:
            print(True)
    else:
        print(False)
        
my_func([1,2,4,0,0,7,5])     

#12. Write a function that returns the number of prime numbers that exist up to and including a given number

def my_func(a):
    total=0
    for num in range(0,a+1):
        if not(num%2==0 or num%3==0 or num%5==0 or num%7==0):
            total+=1
    print(total+3)
        
my_func(200)     

```
